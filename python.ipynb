```python
#extract all these columns using for loop
album_elements = []
for row in data["items"]:
  album_id = row["track"]["album"]["id"]
  album_name = row["track"]["album"]["name"]
  album_release_date = row["track"]["album"]["release_date"]
  album_total_tracks = row["track"]["album"]["total_tracks"]
  album_url = row["track"]["album"]["external_urls"]["spotify"]

  album_element = {"album_id":album_id,"album_name":album_name,
                  "album_release_date":album_release_date,"album_total_tracks":album_total_tracks,"album_url":album_url}
  album_elements.append(album_element)

album_elements
```

```python
artist_list = []
for i in data["items"]:
  for k,v in i.items():
    if k == 'track':
      for artist in v["artists"]:
        artist_dict = {"artist_id":artist["id"],"artist_name":artist["name"]}
        artist_list.append(artist_dict)
artist_list
```
